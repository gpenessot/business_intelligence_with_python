Ce code Python utilise Matplotlib pour créer un graphique combiné affichant les nuitées et la température sur une période de 35 jours. Contrairement à la version précédente, ce code utilise un seul axe principal avec un axe secondaire superposé. Le graphique à barres grises pour les nuitées est créé avec ax1.bar() sur l'axe principal, tandis que la courbe rouge pour la température est tracée avec ax2.plot() sur un axe secondaire créé par ax1.twinx(). Cette approche permet de superposer directement les deux séries de données. Les étiquettes des axes sont définies séparément pour chaque axe avec set_xlabel() et set_ylabel(). La légende est construite en combinant les éléments des deux axes, ce qui permet d'avoir une légende unifiée. L'utilisation de plt.tight_layout() assure un ajustement automatique de la mise en page. Cette méthode crée un graphique plus compact où les deux séries de données partagent le même espace, facilitant la comparaison directe entre les nuitées et la température.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
nights = [515, 420, 370, 250, 135, 120, 60, 20]
temp = [29, 23, 27, 25, 20, 23, 23, 17]
days = np.arange(0, 40, 5)  # 0, 5, 10, ..., 35

# Create the figure and axes
fig, ax1 = plt.subplots(figsize=(12, 6))

# Bar chart for nights
ax1.bar(days, nights, align='center', alpha=0.8, color='grey', label='Nights', width=4)
ax1.set_xlabel('Days')
ax1.set_ylabel('Number of nights')

# Create a second y-axis for temperature
ax2 = ax1.twinx()

# Line plot for temperature
ax2.plot(days, temp, color='red', marker='o', linewidth=2, label='Temperature')
ax2.set_ylabel('Temperature (°C)')

# Title of the chart
plt.title('Nights and Temperature over a 35-Day Period')

# Legend
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper right')

# Adjust layout
plt.tight_layout()
fig.savefig("./images/ch07_32.png")
# Display the chart
plt.show()
```

Ce code Python utilise la bibliothèque Matplotlib pour créer un graphique combiné présentant l'évolution des nuitées et de la température sur une période de 35 jours. Le graphique est divisé en deux parties superposées : une courbe rouge en haut pour la température et un diagramme à barres grises en bas pour les nuitées.

La figure est créée avec plt.subplots(), qui génère deux sous-graphiques verticalement alignés avec un ratio de hauteur de 2:3. La courbe de température est tracée avec ax1.plot(), tandis que les barres des nuitées sont créées avec ax2.bar(). Les axes superflus sont supprimés pour un rendu épuré, et les valeurs sont directement annotées sur le graphique avec ax1.annotate() pour les températures et ax2.text() pour les nuitées. 

Le partage de l'axe x (sharex=True) et l'ajustement de l'espacement (fig.subplots_adjust(hspace=0)) assurent un alignement parfait entre les deux graphiques. Enfin, les titres et étiquettes sont ajoutés pour une meilleure compréhension du graphique.

```python
# Create the figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 6), height_ratios=[1, 3], sharex=True)
fig.subplots_adjust(hspace=0)  # Remove space between subplots

# Temperature plot (top)
ax1.plot(days, temp, color='red', marker='o', linewidth=2, label='Temperature')
ax1.set_ylim(min(temp) - 5, max(temp) + 5)  # Adjust y-axis limits
ax1.spines['top'].set_visible(False)
ax1.spines['right'].set_visible(False)
ax1.spines['left'].set_visible(False)
ax1.spines['bottom'].set_visible(False)  # Remove x-axis
ax1.yaxis.set_visible(False)
ax1.xaxis.set_visible(False)  # Remove x-axis ticks

# Add values on the temperature curve points
for i, txt in enumerate(temp):
    ax1.annotate(f'{txt}°C', (days[i], temp[i]), xytext=(0, 5), 
                 textcoords='offset points', ha='center', va='bottom',
                 color='red', fontweight='bold')

# Bar chart for nights (bottom)
bar_width = 3.5
bars = ax2.bar(days, nights, align='center', alpha=0.8, color='grey', label='Nights', width=bar_width)

# Remove y-axis for the bar chart
ax2.yaxis.set_visible(False)
ax2.spines['left'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Add values on the bars
for block in bars:
    height = block.get_height()
    ax2.text(block.get_x() + block.get_width()/2., height,
             f'{height}',
             ha='center', va='bottom')

# x-axis label (only for the bottom chart)
ax2.set_xlabel('Days')

# Title of the chart
fig.suptitle('Nights and Temperature over a 35-Day Period', y=0.95)

# Adjust layout
plt.tight_layout()
fig.savefig("./images/ch07_33.png")
# Display the chart
plt.show()
```
